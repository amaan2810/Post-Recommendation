```python
# Import the libraries

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read the data

views = pd.read_csv('E:/Work/Amaan/Jobs/SocialComment/views.csv')
views.head()
```

```python
posts = pd.read_csv('E:/Work/Amaan/Jobs/SocialComment/posts.csv')
posts.head()
```

```python
# Join the dataframes so we can know which users have seen which posts

view_posts = pd.merge(views, posts, left_on='post_id', right_on='_id')
view_posts.head()
```

```python
# Posts by number of views

view_stats = pd.DataFrame(view_posts.groupby('title')['timestamp'].count().reset_index(name='views').sort_values(['views'], ascending=False))
view_stats = view_stats.set_index('title')
view_stats.head(10)
```

```python
view_stats.shape
```

```python
# Add a flag so that we know which user has seen which posts

view_posts['Flag'] = np.nan
view_posts['Flag'] = view_posts['Flag'].fillna(1)
view_posts.head()
```

```python
# User & title pivot table

P = view_posts.pivot_table(index='user_id', columns='title', values='Flag', aggfunc=np.sum, fill_value=0)
P
```

```python
# Function to create post wise recommendations

def get_post_wise_recommendations(post_title):
    # Get the similarity correlated scores of all posts with that post
    sim_scores = P.corrwith(P[post_title])

    # Sort the posts based on the similarity scores
    corr_title = pd.DataFrame(sim_scores, columns=['Correlation'])
    
    # Removing na values 
    corr_title.dropna(inplace=True)
    
    # Join the stats to see number of views
    corr_title = corr_title.join(view_stats['views'])
    
    # Remove the mentioned post so that it does not recommend itself
    corr_title = corr_title.drop([post_title])
    
    # Return the top 10 similar posts
    return corr_title.sort_values('Correlation', ascending=False).head(10)
```

```python
# Find the cosine similarity between users

cosine = cosine_similarity(P)
np.fill_diagonal(cosine, 0 )
user_similarity = pd.DataFrame(cosine, index=P.index)
user_similarity.columns = P.index
user_similarity.head(10)
```

```python
# User similarity

user_id = '5deeef6142a8854bf6eabab9'
sim_scores = user_similarity.corrwith(user_similarity[user_id])
corr_title = pd.DataFrame(sim_scores, columns=['Correlation'])
corr_title = corr_title.drop([user_id])
corr_title.sort_values('Correlation', ascending=False).head(10)
```

```python
# Function to create user based recommendations

def get_user_recommendation(user):
    sim_scores = user_similarity.corrwith(user_similarity[user])
    corr_title = pd.DataFrame(sim_scores, columns=['Correlation'])
    corr_title = corr_title.drop([user])
    top3_sim_users = corr_title.sort_values('Correlation', ascending=False).head(3)
    
    # top 3 users
    user1 = top3_sim_users.index[0]
    user2 = top3_sim_users.index[1]
    user3 = top3_sim_users.index[2]
    
    # posts viewed by top 3 and mentioned user
    user_posts = P[((P.index == user) == True)]
    user1_posts = P[((P.index == user1) == True)]
    user2_posts = P[((P.index == user2) == True)]
    user3_posts = P[((P.index == user3) == True)]
    
    # user recommendations
    user_recommendations = []
    
    # Recommend posts which user has not seen and user1 has seen
    for post1 in user_posts:
        if(user_posts[post1].values == 0):
            if(user1_posts[post1].values != 0):
                user_recommendations.append(post1)
    
    # Recommend posts which user has not seen and user2 has seen
    for post2 in user_posts:
        if(user_posts[post2].values == 0):
            if(user2_posts[post2].values != 0):
                if post2 not in user_recommendations:
                    user_recommendations.append(post2)
    
    # Recommend posts which user has not seen and user3 has seen
    for post3 in user_posts:
        if(user_posts[post3].values == 0):
            if(user3_posts[post3].values != 0):
                if post3 not in user_recommendations:
                    user_recommendations.append(post3)
                    
    # Return the list of recommendations
    return [post for post in user_recommendations]
```

```python
# Function to create top recommendation for a user

def get_top_user_recommendation(user):
    sim_scores = user_similarity.corrwith(user_similarity[user])
    corr_title = pd.DataFrame(sim_scores, columns=['Correlation'])
    corr_title = corr_title.drop([user])
    top3_sim_users = corr_title.sort_values('Correlation', ascending=False).head(3)
    
    # top 3 users
    user1 = top3_sim_users.index[0]
    user2 = top3_sim_users.index[1]
    user3 = top3_sim_users.index[2]
    
    # posts viewed by top 3 and mentioned user
    user_posts = P[((P.index == user) == True)]
    user1_posts = P[((P.index == user1) == True)]
    user2_posts = P[((P.index == user2) == True)]
    user3_posts = P[((P.index == user3) == True)]
    
    # user recommendations
    user_recommendations = []
    
    # Recommend posts which user has not seen and user1, user2 and user3 have seen
    for post in user_posts:
        if(user_posts[post].values == 0):
            if(user1_posts[post].values != 0):
                if(user2_posts[post].values != 0):
                    if(user3_posts[post].values != 0):
                        user_recommendations.append(post)

    # Return the list of recommendations
    return [post for post in user_recommendations]
```
